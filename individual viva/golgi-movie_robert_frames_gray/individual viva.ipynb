```python
import glob
file_name = glob.glob('golgi-movie_robert_frame_*_gray.jpg')

file_names = sorted(file_name)

print(file_names)
```

```python
sum_list = []

for files in file_names:
    all_files = plt.imread(files)
    sums = np.sum(all_files)
    sum_list.append(sums)
    
print(sum_list)
```

```python
my_max = 0

for sums in sum_list:
    if sums > my_max:
        my_max = sums

print(my_max)
```

```python
total_fluorescence_signal = []

for sums in sum_list:
    signal = sums/my_max
    total_fluorescence_signal.append(signal)
    
print(total_fluorescence_signal)
    
```

```python
from matplotlib import pyplot as plt

x = [1,18,1]
y = total_fluorescence_signal

plt.errorbar(x, y, color='blue', linestyle='dashed',
             marker='o', markerfacecolor='blue')

plt.xlabel('Frame Number')
plt.ylabel('Total fluorescence signal')

plt.show
```
